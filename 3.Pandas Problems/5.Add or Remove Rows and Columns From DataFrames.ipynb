```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'first':['saurav','gaurav','aadesh'],
                   'last':['singh','singh','dangi'],
                   'email':['sauravsingh@gmail.com','gauravsingh@gmail.com','aadesh@gmail.com']})
```

```python
df
```

```python
#inserting a column 
df.insert(3,'City',['Indore','Bhopal','ujjain'])
```

```python
#adding column
df['full_name'] = df['first'] + " " +df['last']
df
```

```python
#dropping column
# df.drop(['first','last'],axis=1,inplace=True)
df.drop(columns=['first','last'],inplace=True)
df
```

```python
#getting column from full name
df[['first','last']] = df['full_name'].str.split(" ",expand=True)
```

```python
df
```

```python
#adding one row
df.append({'first':'hitakshi',
           'last':'chellani',
           'email':'hitakshi@yahoo.com'},ignore_index=True)
```

```python
#addind another df
df2 = pd.DataFrame({'first':['aayu','mark'],
                   'last':['jain','zuker'],
                   'email':['aayujain@gmail.com','mark@gmail.com'],
                   })
df= df.append(df2,ignore_index=True,sort=False)
df
```
